```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("G:/Big Data/Data/Maggi_TwitterData.csv",encoding = "ISO-8859-1")
```

```python
data.head()
```

```python
data["Tweet"] = data["Tweet"].astype(str)
data["Tweet"].dtypes
```

```python
finaltweets = ""
for eachtweet in data["Tweet"]:
    finaltweets += eachtweet
#len(finaltweets)
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color="white",
                          width=900,
                          height=700
                         ).generate(finaltweets)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
